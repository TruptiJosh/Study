```python
#Q4. Write a simple class to calculate simple interest by taking all three inputs by user in inputdata function and also call
#calculatesimple interest function from input data function.
```

```python
class SimpleInterestCalculator :
    def __init__ (self):
        #initialize variables to store principal , rate and time 
        self.principal = 0
        self.rate=0
        self.time=0

    def input_data (self):
        #take inputs from user
        self.principal = float(input("Enter the principal amount:"))
        self.rate= float(input ("Enter the rate of interest in %:"))
        self.time= float(input("Enter the time in years :"))

        #call the function
        interest = self.calculate_simple_interest()
        print (f"Calculated Simple Interest: {interest}")

        compound_interest = self.calculate_compound_interest()
        print(f"Compound Interest: {compound_interest:.2f}")

        # Calculate and display EMI
        emi = self.calculate_emi()
        print(f"EMI (Monthly Installment): {emi:.2f}")


    def calculate_simple_interest (self):
        #SI =(P*R*T)/100
        return (self.principal * self.rate * self.time)/100

    def calculate_compound_interest(self):
        # Formula for Compound Interest: CI = P * (1 + R/100)^T - P
        return self.principal * ((1 + self.rate / 100) ** self.time) - self.principal

    def calculate_emi(self):
        # EMI formula: EMI = [P * R * (1 + R)^N] / [(1 + R)^N - 1]
        # Convert annual rate to monthly rate
        monthly_rate = self.rate / (12 * 100)
        # Total number of monthly installments
        months = self.time * 12
        # EMI calculation
        emi = (self.principal * monthly_rate * (1 + monthly_rate) ** months) / ((1 + monthly_rate) ** months - 1)
        return emi



#Example Usage
if __name__ == "__main__":
    calculator= SimpleInterestCalculator()
    calculator.input_data()

   
```

```python
from scipy.stats import norm

def calculate_probability_within_range(mean, std_dev, lower_bound, upper_bound):
    # Calculate the CDF for the upper and lower bounds
    cdf_upper = norm.cdf(upper_bound, loc=mean, scale=std_dev)
    cdf_lower = norm.cdf(lower_bound, loc=mean, scale=std_dev)
    
    # The probability is the difference between the two CDF values
    probability = cdf_upper - cdf_lower
    
    return probability



calculate_probability_within_range(100,15,90,110)

```

```python
#Data Preprocessing
#Drop unnecessary columns (eg. dates ,venue if irrelevent)
df=df.drop(columns=['Date','Venue','Column1'])
df
```

```python
# Encoding Categorical Variables
encoder = LabelEncoder()
categorical_columns = ['Out/Not Out', 'Against', 'Format', 'Man of the Match', 'Captain', 'Result']
for col in categorical_columns:
    df[col] = encoder.fit_transform(df[col])

# Splitting the Data
X = df.drop(columns=['Result'])  # Features
y = df['Result']  # Target
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Model Pipeline and Hyperparameter Tuning
pipeline = Pipeline([
    ('Classifier', RandomForestClassifier(random_state=42))
])

param_grid = {
    'Classifier__n_estimators': [10, 50, 100],
    'Classifier__max_depth': [None, 10, 20]
}

grid_search = GridSearchCV(pipeline, param_grid, cv=2, scoring='accuracy', n_jobs=-1)
grid_search.fit(x_train, y_train)

# Best Parameters and Model Evaluation
print("Best Parameters:", grid_search.best_params_)
best_model = grid_search.best_estimator_

# Predictions and Evaluation
y_pred = best_model.predict(x_test)
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Save the Trained Model
joblib.dump(best_model, "virat_kohli_result_model.pkl")
print("\nModel saved as 'virat_kohli_result_model.pkl'")
```
