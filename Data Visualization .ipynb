```python
sal=[15,22,24,30,23,34,43,19,20,52,21]
sal.sort()
sal
```

```python
# 25% percentile = 0.25*(11+1)
```

```python
0.25*12  #ascending 3rd number i.e. 20 is 25th percentile value for this dataset.
```

```python
#3rd position is 25% percentile position for box plot
df1=pd.DataFrame(data=sal)
df1.head(11)
```

```python
df1.plot(kind='box',figsize=(8,5),color='r',yticks=range(0,55,5))
```

```python
#Outliers
import pandas as pd
ser=pd.Series(sal)
ser
```

```python
#1stquantile
q1=ser.quantile(0.25)
print('Q1',q1)
```

```python
#3rdquantile
q3=ser.quantile(0.75)
print('Q3',q3)
```

```python
#inter quantile range
iqr=q3-q1
iqr
```

```python
higher=q3+(1.5*iqr)  #data above 49.25 values i.e.52 data of index no 10 is a outlier data of higher side
print(higher)
```

```python
lower=q1-(1.5*iqr) #no outlier data is present on lower side. which are less than 3.25 value.
print(lower)
```

**#Scatter_matrix**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


from pandas.plotting import scatter_matrix
```

```python
iris=sns.load_dataset('iris')
df=iris.drop(['species'],axis=1)
df.head()
```

```python
#kde=kernel density estimation
scatter_matrix(df,figsize=(10,8),diagonal='kde',color='b')
plt.show()
```

```python
#without kde=kernel density estimation
scatter_matrix(df,figsize=(10,8),color='b')
plt.show()
```
